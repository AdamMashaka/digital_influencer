```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_json('/kaggle/input/synthetic-influencer-backstory/influencer_data.jsonl', lines=True)
```

```python
# Basic information about the dataset
print("üîç Basic Information of Dataset üîç")
print(df.info())
```

```python
# Descriptive statistics of the dataset
print("\nüìä Descriptive Statistics üìä")
print(df.describe())
```

```python
# Distribution of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('üéÇ Distribution of Age üéÇ')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender count
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df, palette='pastel')
plt.title('üöª Sex Distribution üöª')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Lifestyle distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='Lifestyle', data=df, palette='Set2')
plt.title('üíÉ Lifestyle Distribution üíÉ')
plt.xlabel('Lifestyle')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```
